```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
from statsmodels.tsa.holtwinters import SimpleExpSmoothing # SES
from statsmodels.tsa.holtwinters import Holt # Holts Exponential Smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing 
```

```python
series = pd.read_excel('Airlines+Data.xlsx', header=0, index_col=0, parse_dates=True)
series.head(25)
```

From the above box plot we can see that there are no outliers.

```python
#Decomposing the trend and seasonal components

from statsmodels.tsa.seasonal import seasonal_decompose
ts_decompose = seasonal_decompose(np.array(series['Passengers']),model = 'multiplicative',period = 12)

ts_plot = ts_decompose.plot()
```

The passengers has an upward trend. It also has seasonality and the seasonality index varies from 0.8 to 1.2 over the trend.

### Check for stationarity

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

acf_plot = plot_acf(series.Passengers,lags=20)
```

The acf is decreasing slowly, that means its not stationary.

```python
pacf_plot = plot_pacf(series.Passengers,lags=20)
```

```python
#Dickey-Fuller test for stationarity

from statsmodels.tsa.stattools import adfuller
```

```python
def adfuller_test(ts):
    adfuller_result = adfuller(ts,autolag=None)
    adfuller_out = pd.Series(adfuller_result[0:4], index=['Test Statistic','p=value','Lags used','No. of obvs. used'])
    print(adfuller_out)
```

```python
adfuller_test(series.Passengers)
```

The p-value(0.99) is >0.05, which indicates the series is not stationary. So, we have to make it stationary before fitting a model.

```python
#Differencing to attain stationarity

series['diff'] = series.Passengers - series.Passengers.shift(1)
series.head()
```

```python
#Dropping the nan 
series_diff = series.dropna()
```

```python
#plot the first difference

plt.figure(figsize=(10,4))
plt.xlabel('Date')
plt.ylabel('First order Difference')
plt.plot(series_diff['diff'])
```

This plot indicates multiplicative seasonality.

```python
acf_plot = plot_acf(series['diff'].dropna())
```

```python
series.shape
```

### Splitting Data

```python
#train and test data

train = series[0:70]
test = series[70:]
```

### Model 1 - ARIMA

```python
#Arima model fit
from statsmodels.tsa.arima_model import ARIMA

arima = ARIMA(train.Passengers.to_numpy(),order = (1,1,1))
arima_model = arima.fit()
arima_model.summary2()
```

The p values are significant.

```python
#ACF of Residuals
acf_plot = plot_acf(arima_model.resid,lags=20)
```

The residuals doesnot show any significant aotucorrelations, so the model can be assumed to be valid.

```python
#forecasting the future values of 12 quarters

passen_predict, stderr, ci = arima_model.forecast(steps=26)
passen_predict
```

```python
arima_model.plot_predict(1, 96)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error

rmse = sqrt(mean_squared_error(test['Passengers'], passen_predict))
rmse
```

### Model 2 - simple exponential smoothing

```python
ses_model = SimpleExpSmoothing(train["Passengers"]).fit(smoothing_level=0.2,)
model2 = ses_model.forecast(26)
rmse = sqrt(mean_squared_error(test['Passengers'], model2))
rmse
```

### Model 3 - Holt

```python
holt_model = Holt(train["Passengers"]).fit(smoothing_level=0.8,smoothing_trend = 0.2)
model3 = holt_model.forecast(26)
rmse = sqrt(mean_squared_error(test['Passengers'], model3))
rmse
```

### Model 4 - Exponential smoothing with additive seasonality and trend

```python
hwe_model_add = ExponentialSmoothing(train["Passengers"],seasonal="add",trend="add",seasonal_periods=4).fit()
model4 = hwe_model_add.forecast(26)
rmse = sqrt(mean_squared_error(test['Passengers'], model4))
rmse
```

### Model 5 - Exponential smoothing with multiplicative seasonality and additive trend

```python
hwe_model_mul = ExponentialSmoothing(train["Passengers"],seasonal="mul",trend="add",seasonal_periods=4).fit() 
model5 = hwe_model_mul.forecast(26)
rmse = sqrt(mean_squared_error(test['Passengers'], model5))
rmse
```

```python
data = {"MODEL":pd.Series(["ARIMA(1,1,1)","SES","Holt","Exp_add","Exp_multi"]),"RMSE_Values":pd.Series([48.65,72.31,130.56,61.57,59.59])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_Values'])
```

Therefore, from the above table we can say that the ARIMA(1,1,1) method has the least rmse. so, that model should be selected.
