```python
#Decomposing the trend and seasonal components

from statsmodels.tsa.seasonal import seasonal_decompose
ts_decompose = seasonal_decompose(np.array(series['Sales']),model = 'multiplicative',period = 4)

ts_plot = ts_decompose.plot()
```

The prices has an upward trend. It also has seasonality and the seasonality index varies from 0.9 to 1.1 over the trend.

### Check for stationarity

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

acf_plot = plot_acf(series.Sales,lags=20)
```

There is autocorrelation till lag 4. So, this is not stationary. It could be a AR(4) model.

```python
pacf_plot = plot_pacf(series.Sales,lags=20)
```

For lag 1 the partial autocorrelation is significant.

```python
#Dickey-Fuller test for stationarity

from statsmodels.tsa.stattools import adfuller
```

```python
def adfuller_test(ts):
    adfuller_result = adfuller(ts,autolag=None)
    adfuller_out = pd.Series(adfuller_result[0:4], index=['Test Statistic','p=value','Lags used','No. of obvs. used'])
    print(adfuller_out)
```

```python
adfuller_test(series.Sales)
```

The p-value(0.98) is >0.05, which indicates the series is not stationary. So, we have to make it stationary before fitting a model.

```python
#Differencing to attain stationarity

series['Sales_diff'] = series.Sales - series.Sales.shift(1)
series.head()
```

```python
#Dropping the nan 
series_diff = series.dropna()
```

```python
#plot the first difference

plt.figure(figsize=(10,4))
plt.xlabel('Date')
plt.ylabel('First order Difference')
plt.plot(series_diff.Sales_diff)
```

This plot clearly indicates multiplicative seasonality.

```python
acf_plot = plot_acf(series.Sales_diff.dropna())
```

The Acf plot immediately cuts off to zero, thus showing stationarity.

##  Splitting Data

```python
#train and test data

series_train = series[0:30]
series_test = series[30:]
```

### Model 1 - ARIMA

```python
#Arima model fit
from statsmodels.tsa.arima_model import ARIMA

arima = ARIMA(series_train.Sales.to_numpy(),order = (1,1,1))
arima_model = arima.fit()
arima_model.summary2()
```

```python
#ACF of Residuals
acf_plot = plot_acf(arima_model.resid,lags=20)
```

```python
pacf_plot = plot_pacf(arima_model.resid,lags=10)
```

The residuals doesnot show any significant aotucorrelations, so the model can be assumed to be valid.

```python
#forecasting the future values of 12 quarters

sales_predict, stderr, ci = arima_model.forecast(steps=12)
sales_predict
```

```python
arima_model.plot_predict(1, 42)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error

rmse = sqrt(mean_squared_error(series_test['Sales'], sales_predict))
rmse
```

### Model 2 - Linear model

```python
#create new variable t for Quarter

series['t'] = range(1,43)
series.head()
```

```python
#train and test data

train = series[0:30]
test = series[30:]
```

```python
linear_model = sm.OLS(train.Sales,train.t).fit()
pred_linear =  pd.Series(linear_model.predict(pd.DataFrame(test['t'])))
rmse_linear = np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_linear))**2))
rmse_linear
```

### Model 3 - simple exponential smoothing

```python
ses_model = SimpleExpSmoothing(train["Sales"]).fit(smoothing_level=0.2,)
model2 = ses_model.forecast(12)
rmse = sqrt(mean_squared_error(test['Sales'], model2))
rmse
```

### Model 4 - Holt

```python
holt_model = Holt(train["Sales"]).fit(smoothing_level=0.8,smoothing_trend = 0.2)
model3 = holt_model.forecast(12)
rmse = sqrt(mean_squared_error(test['Sales'], model3))
rmse
```

### Model 5 - Exponential smoothing with additive seasonality and trend

```python
hwe_model_add = ExponentialSmoothing(train["Sales"],seasonal="add",trend="add",seasonal_periods=4).fit()
model4 = hwe_model_add.forecast(12)
rmse = sqrt(mean_squared_error(test['Sales'], model4))
rmse
```

### Model 6 - Exponential smoothing with multiplicative seasonality and additive trend

```python
hwe_model_mul = ExponentialSmoothing(train["Sales"],seasonal="mul",trend="add",seasonal_periods=4).fit() 
model5 = hwe_model_mul.forecast(12)
rmse = sqrt(mean_squared_error(test['Sales'], model5))
rmse
```

```python
data = {"MODEL":pd.Series(["ARIMA(1,1,1)","Linear","SES","Holt","Holt_add","Holt_multi"]),"RMSE_Values":pd.Series([583.27,854.13,1153.70,559.52,245.48,181.70])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_Values'])
```

Therefore, from the above table we can say the Holt method with multiplicative seasonality and additive trend has the least rmse. so, that model should be selected.
