```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score,roc_curve
from sklearn.metrics import confusion_matrix
```

```python
#import data
company = pd.read_csv('Company_Data.csv')
company.head()
```

```python
company.info()
```

```python
company.describe()
```

## Data Preprocessing

```python
#encoding the categorical variables
encoder = preprocessing.LabelEncoder()
company['ShelveLoc'] = encoder.fit_transform(company['ShelveLoc'])
company['Urban'] = encoder.fit_transform(company['Urban'])
company['US'] = encoder.fit_transform(company['US'])

company.head()
```

```python
#Encoding the sales column by setting values greter than mean(7.5) to 1 and less than mean to 0.
company['sales_encode']=''
company.loc[company['Sales']>7.5,'sales_encode']='1'
company.loc[company['Sales']<=7.5,'sales_encode']='0'

company.head()
```

```python
company.drop('Sales',axis=1,inplace = True)
```

```python
company['sales_encode'].value_counts()
```

#### The data is balanced.

```python
company.head()
```

```python
Y = company.iloc[:,-1]
X = company.iloc[:,0:-1]
```

```python
# Splitting data into training and testing data set
x_train, x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2,random_state=40)
```

## Visualise Decision Trees

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf_tree_entropy,
               feature_names = x_train.columns, 
               class_names=['0 or Good','1 or Risk'],
               filled = True,rounded=True);
```

## Visualization using Graphviz

```python
import graphviz 
import pydotplus as pdot
from IPython.display import Image

dot_data = tree.export_graphviz(clf_tree_entropy, out_file="companyDT.odt", 
                              feature_names=x_train.columns,  
                              class_names=['0 or Less Sales','1 or More Sales'],  
                              filled=True, rounded=True,  
                              special_characters=True)

graph = graphviz.Source(dot_data) 

graph = pdot.graphviz.graph_from_dot_file('companyDT.odt')
graph.write_jpg('companyDT.png')

Image(filename = 'companyDT.png')
```

we can find out a few rules of identifying the correct class from the decision tree.

2. when Price is less than 131.5 , advertising cost is less than 6.5,age is less than 60.5,shelve location is not good,factory is not in urban location, then the company will make more sales.

Similarly, we can find out more rules to identify the class labels.

## Measuring Test Acuuracy

```python
y_pred = clf_tree_entropy.predict(x_test)
roc_auc_score(y_test,y_pred)
```

### Confusion Matrix

```python
confusion_matrix(y_test, y_pred)
```

### Classification Report

```python
print(classification_report(y_test, y_pred))
```

## Final Conclusions

1. In the model with criterion entropy index, the training-set accuracy score is 0.79 while the test-set accuracy to be 0.72. These two values are quite comparable. So, there is no sign of overfitting.
2. From the confusion matrix we see that it predicts both the classes but still there are type1 and type 2 errors.
3. From the classification report we see that the recall for 0 is 0.78 and for 1 it is 0.66. 
4. Overall we can say that this model perfoms well in predicting the class levels.
