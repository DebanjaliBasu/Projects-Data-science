```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
#importnig train data
filename = ('forestfires.csv')
df = pd.read_csv(filename)
```

```python
##Dropping the month and day columns as we already have the dummy values for day and month in following columns
df.drop(["month","day"],axis=1,inplace =True)
df.head()
```

```python
#splitting data into x and y

x = df.iloc[:,1:-1]
y = df.iloc[:,-1]
```

```python
#Normalizing data

def norm_func(i):
    X = (i-i.min())/(i.max()-i.min())
    return (X)
x = norm_func(x)
```

```python
# dividing data into train and test 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20)
```

```python
x_train.head()
```

## Grid Search CV 

#### fitting gridsearch model to find out optimum values for SVM

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,0.5],'C':[10,1,0.001]}]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

### Fitting the SVM model now with grid search results for best model

```python
clf = SVC(C = 0.001, gamma = 50,kernel = 'rbf')
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test,y_pred)*100
print("Accuracy = ", acc)
confusion_matrix(y_test,y_pred)
```
