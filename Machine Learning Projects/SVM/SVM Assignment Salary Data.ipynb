```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
#importnig train data
df = pd.read_csv('SalaryData_Train(1).csv')
df.head()
```

```python
## Label encoding  train data

for i in df.columns:
    df[i] = LabelEncoder().fit_transform(df[i])
```

```python
# dividing data into X and Y
x_train = df.iloc[:,1:-1]
y_train = df.iloc[:,-1]
```

```python
#importing test data
df_test = pd.read_csv('SalaryData_Test(1).csv')
df_test.head()
```

```python
# Label encoding test data
for i in df_test.columns:
    df_test[i]= LabelEncoder().fit_transform(df_test[i])
```

```python
# dividing test data into X and Y
x_test = df_test.iloc[:,1:-1]
y_test = df_test.iloc[:,-1]
```

```python
#Normalizing data

def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
x_train = norm_func(x_train)
x_test =  norm_func(x_test)
y_train = norm_func(y_train)
y_test =  norm_func(y_test)
```

```python
x_train.head()
```

## Grid Search CV 

#### fitting gridsearch model to find out optimum values for SVM

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,0.5],'C':[10,1,0.001]}]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

### Fitting the SVM model now with grid search results for best model

```python
clf = SVC(C = 1, gamma = 5,kernel = 'rbf')
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test,y_pred)*100
print("Accuracy = ", acc)
confusion_matrix(y_test,y_pred)
```
