```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
#inport data
train = pd.read_csv("SalaryData_Train.csv")
test = pd.read_csv("SalaryData_Test.csv")
```

```python
train.head()
```

## Data Preprocessing

```python
df_train = pd.get_dummies(train)
df_train.head()
```

```python
df_train.info()
```

```python
X = df_train.iloc[:,0:-2]
y = df_train.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
#standardising data
x = StandardScaler().fit_transform(X)
```

```python
x
```

## Model Fitting

```python
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(x, y)
```

```python
test.head()
```

```python
df_test= pd.get_dummies(test)
df_test.head()
```

```python
X_test = df_test.iloc[:,0:-2]
y_test = df_test.iloc[:,-1]
```

```python
y_predict = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_predict)
```

#### The Gaussian Naive bayes model gives 77.7% accuracy.

```python
#Multinomial Naive Bayes

from sklearn.naive_bayes import MultinomialNB as MB

clf_mb = MB()
clf_mb.fit(X,y)
```

```python
y_pred = clf_mb.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

#### The accuracy of Multinomial Naive Bayes model is 77.49% .
